```python
import pandas as pd
file_path = 'data/Seamen Report_rev1.xlsx'
xls = pd.ExcelFile(file_path)

# Load sheets
#container_captain_df = pd.read_excel(xls, sheet_name='Container-Captain')
#manalagi_captain_df = pd.read_excel(xls, sheet_name='Manalagi-Captain')
container_KKM_df = pd.read_excel(xls, sheet_name='Container-KKM')
manalagi_KKM_df = pd.read_excel(xls, sheet_name='Manalagi-KKM')

```

```python
import pandas as pd


def prepare_display_df(df):
    display_df = df[['SEAMAN CODE', 'SEAMAN NAME', 'VESSEL GROUP ID', 'RANK', 'CERTIFICATE']].copy() 
    return display_df

def prioritize_nakhoda_ant1(df):
    df['priority'] = ((df['RANK'] == 'KKM') & (df['CERTIFICATE'] == 'ATT-I')).astype(int) * 2
    df['priority'] += (df['CERTIFICATE'] == 'ATT-I').astype(int)  
    sorted_df = df.sort_values(by='priority', ascending=False).drop(columns=['priority'])
    return sorted_df

def filter_group_1(df):
    filtered_df = df[df['VESSEL GROUP ID'] == 1].copy()
    return filtered_df

container_captain_filtered = prepare_display_df(container_KKM_df)
manalagi_captain_filtered = prepare_display_df(manalagi_KKM_df)

container_captain_prioritized = prioritize_nakhoda_ant1(container_captain_filtered)
manalagi_captain_prioritized = prioritize_nakhoda_ant1(manalagi_captain_filtered)

container_captain_group1 = filter_group_1(container_captain_prioritized)
manalagi_captain_group1 = filter_group_1(manalagi_captain_prioritized)
```
